```python
import pandas as pd 
import numpy as np
import neattext.functions as nfx
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

from sklearn.metrics.pairwise import cosine_similarity
```

```python
df= pd.read_csv('udemy_course_data.csv')
df.head()
```

```python
df['Clean_title'] = df['course_title'].apply(nfx.remove_stopwords)

df['Clean_title'] = df['Clean_title'].apply(nfx.remove_special_characters)  

df.head()
```

## Vectorizing the Clean Title

```python
countvect = CountVectorizer()
cvmat = countvect.fit_transform(df['Clean_title'])
cvmat
```

## Cosine Similarity

```python
cos_sim = cosine_similarity(cvmat)  
cos_sim
```

## Recommended Course

```python
title = "Learn jQuery from Scratch - Master of JavaScript library"
```

```python
course_index = pd.Series(
    df.index, index= df['course_title']).drop_duplicates()

course_index[title]
```

```python
index = course_index[title]
scores= list(enumerate(cos_sim[index]))
scores
```

```python
course_index = pd.Series(
    df.index, index= df['course_title']).drop_duplicates()

index= course_index[title]

scores= list(enumerate(cos_sim[index]))

sorted_socres = sorted(scores, key= lambda x: x[1], reverse= True)  

selected_course_index = [i[0] for i in sorted_socres[1:]]

selected_course_score = [i[1] for i in sorted_socres[1:]]

rec_df = df.iloc[selected_course_index]

rec_df['Similiarity_score'] = selected_course_score

final_recommendation = rec_df[['course_title', 'Similiarity_score', 'url', 'price', 'num_subscribers']]

final_recommendation
```

```python
final_recommendation[final_recommendation['Similiarity_score'] > 0.6]
```
